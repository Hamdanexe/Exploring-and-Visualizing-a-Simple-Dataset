```python
import pandas as pd
import matplotlib.pyplot as plt

iris = pd.read_csv("sample_iris.csv")

print("Shape:", iris.shape)
print("Columns:", iris.columns.tolist())
print(iris.head())
print(iris.info())
print(iris.describe())

plt.figure(figsize=(8, 6))
species_unique = iris["species"].unique()
colors = ['r', 'g', 'b']
for i, species in enumerate(species_unique):
    subset = iris[iris["species"] == species]
    plt.scatter(subset["sepal_length"], subset["sepal_width"], 
                color=colors[i % len(colors)], label=species)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Scatter Plot: Sepal Length vs Sepal Width")
plt.legend()
plt.show()

iris.drop(columns=["species"]).hist(figsize=(10, 8), bins=10, color="skyblue", edgecolor="black")
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()

plt.figure(figsize=(10, 6))
iris.drop(columns=["species"]).boxplot()
plt.title("Box Plots of Iris Features")
plt.show()
```
